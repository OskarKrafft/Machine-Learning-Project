```python
# Libraries
import numpy as np
import matplotlib.pyplot as plt

# Data
mcc = [0.806, 0.804, 0.803, 0.796, 0.638]
x = [0, 1, 2, 3, 4]

labels = ("Random Forest", "XGBoost", "Log-Regression", "SVM", "Naïve Bayes")

#Figure
plt.figure(figsize=(10, 7))
font = {'family' : 'normal',
        'weight' : 'normal',
        'size'   : 16}

plt.rc('font', **font)

# Create bar plot
plt.bar(x, mcc, width = 0.6, color=['royalblue', 'royalblue', 'royalblue', 'royalblue', 'royalblue'])
plt.ylim([0.5, 0.85])
plt.xticks([0, 1, 2, 3, 4], labels)

# Create axis label and title
plt.ylabel("f1")
#plt.title("Model performance (f1 Score)")


# Show graphic

plt.savefig("f1_model_comparison_title", bbox_inches = "tight", dpi = 144)
```
